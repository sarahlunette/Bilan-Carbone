```python
#Packages
import numpy as np
import matplotlib as plt
import scipy as sp
import pandas as pd
```

```python
#Data
path = '/Users/sarahlenet/Documents/Candidatures Postes/carbon/'
file = 'Données Etude de cas.xlsx'
read_me = pd.read_excel(path + file, sheet_name=0)
products = pd.read_excel(path + file, sheet_name=1)
products_parts = pd.read_excel(path + file, sheet_name=2)
materials = pd.read_excel(path + file, sheet_name=3)
```

```python
products = products.rename(columns = {"id": 'product_id'})
products.index = products.product_id
products
```

```python
products_parts
```

```python
materials = materials.rename(columns = {'id': 'material_id'})
materials.index = materials['material_id']
materials.drop('material_id', inplace = True, axis = 1)
materials
```

```python
#Table joins
products_and_parts = products_parts.join(products, on = 'product_id', rsuffix = '_parts', lsuffix = '_products')
products_and_parts.index = products_and_parts.id
data = products_and_parts.join(materials, on = 'material_id', lsuffix = '_parts', rsuffix = '_materials')
data = data[~data['product_id_products'].isnull()]
data
```

# 

# 

```python
data['Total_weight'] = data['weight']*data['quantity']/1000
data['Total_kgCO2'] = data['Total_weight']*data['value']
data['Max_kgCO2'] = data['Total_kgCO2']*(1 + data['uncertainty']/100)
data['Min_kgCO2'] = data['Total_kgCO2']*(1 - data['uncertainty']/100)
```

```python
data.index = data['id']
data = data.drop(['id'], axis = 1)
```

```python
data
```

```python
data = data.drop(['product_id_products'], axis = 1)
```

```python
data = data.rename(columns = {'name_materials' : 'product_name'})
```

```python
data = data.rename(columns = {'name_parts' : 'container_name'})
```

```python
data
```

```python
data.columns
data['Total_weight'] = data['Total_weight'].apply(round)
data['Total_kgCO2'] = data['Total_kgCO2'].apply(round)
data['Max_kgCO2'] = data['Max_kgCO2'].apply(round)
data['Min_kgCO2'] = data['Min_kgCO2'].apply(round)
```

```python
carbon = data
carbon = carbon[['product_name','product_id_parts','material_id','category','container_name', 'weight', 'quantity', 'unit', 'value', 'Total_weight','Total_kgCO2','uncertainty',  'Max_kgCO2',
       'Min_kgCO2','score_id','coefficient_type','base','source']]
carbon
```

```python
carbon.to_csv(path + 'Carbon.csv')
```

```python
#SQL Request for kgCO2 emission for each product
SELECT SUM(TOTAL_KGCO2) AS sum_carbon, SUM(MAX_KGCO2) AS sum_carbon_max, SUM(MIN_KGCO2) AS sum_carbon_min
FROM(SELECT CONTAINER_NAME, SUM(TOTAL_KGCO2) AS TOTAL_KGCO2, SUM(MAX_KGCO2) AS MAX_KGCO2, SUM(MIN_KGCO2) AS MIN_KGCO2
FROM CARBON
GROUP BY CONTAINER_NAME);
```

# 

# 

```python
carbon_with_uncertainty = carbon[['container_name', 'Total_kgCO2', 'Max_kgCO2', 'Min_kgCO2']].groupby('container_name').sum()
```

```python
carbon_with_uncertainty
```

```python
carbon_with_uncertainty['Diff_low'] = carbon_with_uncertainty['Total_kgCO2'] - carbon_with_uncertainty['Min_kgCO2']
carbon_with_uncertainty['Diff_high'] = carbon_with_uncertainty['Max_kgCO2'] - carbon_with_uncertainty['Total_kgCO2']
carbon_with_uncertainty['uncertainty'] = carbon_with_uncertainty[['Diff_low','Diff_high']].max(axis=1)/carbon_with_uncertainty['Total_kgCO2']*100
carbon_with_uncertainty['uncertainty'] = carbon_with_uncertainty['uncertainty'].apply(round)
carbon_with_uncertainty.drop(['Diff_low','Diff_high'], axis = 1)
```

```python
carbon_with_uncertainty.to_csv(path + 'CO2_uncertainty.csv')
```
